```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

# Load the raster that starts wtih "time_of_arrival_*" in 03-data/outputs
raster_start = "03-data/outputs/time_of_arrival_"
# get any raster file that starts with the above string
import os
raster_files = [f for f in os.listdir("03-data/outputs") if f.startswith("time_of_arrival_")]
if not raster_files:
    raise FileNotFoundError("No raster files found starting with 'time_of_arrival_' in 03-data/outputs")
raster_path = os.path.join("03-data/outputs", raster_files[0])
with rasterio.open(raster_path) as src:
    data = src.read(1)  # Read the first band
# Convert seconds to hours for better readability
data_hours = data / 3600

# Plot
plt.figure(figsize=(10, 8))
plt.imshow(data_hours, cmap='hot_r', origin='upper')
plt.colorbar(label='Time of Arrival (hours)')
plt.title('Fire Spread - Time of Arrival')
plt.axis('off')
plt.show()
```
