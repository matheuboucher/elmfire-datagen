```python
import shutil
import os
import glob

# Source and destination directories
dirs = [
    ('../../data-wildfire/02-multifuel-realistic/processed_sims', './extra-realistic-cases/processed_sims'),
    ('./02-multifuel-realistic/cases', './extra-realistic-cases/cases')
]

for src_path, dst_path in dirs:
    os.makedirs(dst_path, exist_ok=True)
    
    # Get all items and sort by case number
    items = glob.glob(os.path.join(src_path, 'case_*'))
    items.sort(key=lambda x: int(os.path.basename(x).split('_')[1]))
    
    # Move last 500
    for item in items[-500:]:
        item_name = os.path.basename(item)
        shutil.move(item, os.path.join(dst_path, item_name))
        print(f"Moved {item_name}")
```
